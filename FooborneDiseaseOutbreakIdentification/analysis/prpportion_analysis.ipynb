```python
import numpy as np
data=pd.read_csv("../feature_extraction/yisijuji_liudiao_concat_data_without_normed.csv",dtype=object)
data
```

```python
Vibrio_list=pathogen2incident["副溶血性弧菌"]
Salmon_list=pathogen2incident["沙门氏菌"]
Noro_list=pathogen2incident["诺如病毒"]
Diarr_list=pathogen2incident['致泻大肠埃希氏菌属']
#副溶血性弧菌
Vibrio_data = data.loc[data['Incident_GUID']==Vibrio_list[0]]
for i in range(1,len(Vibrio_list)):
    find_data=data.loc[data['Incident_GUID']==Vibrio_list[i]]
    Vibrio_data = pd.concat([Vibrio_data,find_data])
Vibrio_data
```

```python
#沙门氏菌
Salmon_data = data.loc[data['Incident_GUID']==Salmon_list[0]]
for i in range(1,len(Salmon_list)):
    find_data=data.loc[data['Incident_GUID']==Salmon_list[i]]
    Salmon_data = pd.concat([Salmon_data,find_data])
Salmon_data
```

```python
#诺如病毒
Noro_data = data.loc[data['Incident_GUID']==Noro_list[0]]
for i in range(1,len(Noro_list)):
    find_data=data.loc[data['Incident_GUID']==Noro_list[i]]
    Noro_data = pd.concat([Noro_data,find_data])
Noro_data
```

```python
#致泻大肠埃希氏菌属
Diarr_data = data.loc[data['Incident_GUID']==Diarr_list[0]]
for i in range(1,len(Diarr_list)):
    find_data=data.loc[data['Incident_GUID']==Diarr_list[i]]
    Diarr_data = pd.concat([Diarr_data,find_data])
Diarr_data
```

```python
remain_data=data.drop(Vibrio_data.index)
remain_data=remain_data.drop(Salmon_data.index)
remain_data=remain_data.drop(Noro_data.index)
remain_data=remain_data.drop(Diarr_data.index)
remain_data
```

```python
ordered_data=pd.concat([Vibrio_data,Salmon_data])
ordered_data=pd.concat([ordered_data,Noro_data])
ordered_data=pd.concat([ordered_data,Diarr_data])
ordered_data=pd.concat([ordered_data,remain_data])
ordered_data
```

```python
train_data=train_data.fillna('0.0')
train_label=train_data['label'].values
train_data=train_data.drop(['label'],axis=1)
train_data=train_data.drop(['Incident_GUID','num0','num2','false','false.1','false.2'],axis=1)

train_data.columns=col_list3_english
X_train=train_data.values

test_data=test_data.fillna('0.0')
test_label=test_data['label'].values
test_data=test_data.drop(['label'],axis=1)
test_data=test_data.drop(['Incident_GUID','num0','num2','false','false.1','false.2'],axis=1)
X_test=test_data.values
test_data.columns=col_list3_english

train_label=train_label.astype(int)
test_label=test_label.astype(int)

from xgboost import XGBClassifier
clf_xgboost = XGBClassifier(booster='gbtree',objective='binary:logistic',n_estimators=50,learning_rate=0.05,max_depth=4,max_delta_step=2)
clf_xgboost.fit(X_train,train_label)
from sklearn.metrics import  recall_score
y_predict=clf_xgboost.predict(X_test)
print(recall_score(test_label,y_predict))
from sklearn.metrics import f1_score
f1=f1_score(test_label,y_predict)
print(f1)
```
