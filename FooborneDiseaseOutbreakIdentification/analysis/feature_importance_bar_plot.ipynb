```python
plt.figure(figsize=(10,12))
plt.grid(axis="x")
plt.ylabel('Features')
plt.xlabel('Importances')
plt.title('Feature importance of the three models')
x=np.arange(24)#柱状图在横坐标上的位置
#列出你要显示的数据，数据的列表长度与x长度相同
y1_data=df[0:1].values[0].tolist()
y1_data.reverse()
y2_data=df[1:2].values[0].tolist()
y2_data.reverse()
y3_data=df[2:3].values[0].tolist()
y3_data.reverse()

y1=np.array(y1_data)
y2=np.array(y2_data)
y3=np.array(y3_data)
bar_width=0.3#设置柱状图的宽度
colunm_name=['ExPosure_OtherIsill', 'Other Features','Disease_OtherIsill', 'ExPosure_Eattime', 'Disease_SickTime', 'Disease_TreatTime', 'ExPosure_BuyAddress', 'ExPosure_EatAddress', 'ExPosure_EatNum', 'home-made', 'Disease_Address', 'acute gastroenteritis', 'vomiting', 'fungus and its products', 'pre-packaged', 'family', 'Disease_Age', 'poisonous mushroom poisoning', 'farmers', 'catering industry-other', 'ExPosure_EatDistrict', 'food service industry', 'meat and meat products', 'ExPosure_Name']
colunm_name.reverse()
tick_label=colunm_name
#绘制并列柱状图
plt.barh(x-bar_width,y1,bar_width,color='steelblue',label='Random Forest')
plt.barh(x,y2,bar_width,color='salmon',label='Adaboost')
plt.barh(x+bar_width,y3,bar_width,color='green',label='XGBoost')
plt.legend()#显示图例，即label
plt.yticks(x+bar_width/2,tick_label)#显示x坐标轴的标签,即tick_label,调整位置，使其落在两个直方图中间位置
plt.savefig('feature_importances/plot_barh_5.png',dpi=500,bbox_inches = 'tight')
plt.show()
```
